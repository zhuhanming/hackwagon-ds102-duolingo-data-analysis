##### p_recall: What is "lexeme"?

##### timestamp: What is "UNIX timestamp"?

##### lexeme_string: Surface Form?? Lemma?? POS?? Modifiers?!?! What are all these?

<b>Lemma</b><br>
Let us run through all these together. Let us start with lemma. What is lemma? Lemma is basically sort of the "root word", a particular form chosen to represent a lexeme. Let's use the example of `run` from earlier on again. In this example, running, ran, runs, run all have the same <b>core meaning</b> of `run`. To represent this meaning, the particular form "run" has been selected. In other words, run is the root word for ran, running and runs in English, because it is the form that has been selected to represent this deeper concept of `run`. <br>

If that is a bit too confusing, just think of lemma as the accepted root word in a language. Just like eat is the root word for eating, eats, ate. And this brings us to what surface form is. <br>

<b>Surface form</b><br>
Surface form is the final form of a root word after it has been modified due to grammatical reasons. For example, we cannot say `he likes run`, but we can modify the word run to get `he likes running`. In this case, "running" is the surface form.<br>

We also have cases where the surface form is what is called a "wildcard". I've mainly seen 2 cases: `<*sf>` - refers to a "generic" lexeme without any specific surface form aka all forms of a root word is tested, and `<*numb>`, which means both singular and plural forms were tested e.g. teacher and teachers.

<b>POS</b><br>
What is POS then? POS stands for part of speech. Part of speech basically determines what is the function that the word is playing. Some commonly known POS are noun, verb, adjectives. Of course there are many more, but it's basically telling us that the first tag after the lemma is the POS tag. <br>

<b>Modifiers</b><br>
So what are modifiers then? Modifiers are basically ways to describe the surface form. Is it plural? Is it singular? Is it in present tense? For languages that have masculine and feminine words, which one is it? <br>

Some examples for the English language are: `boys/boy<n><pl>` In this example, `boys` is a noun, thus the `<n>` POS tag for noun, and since `boys` is plural, it gets the `<pl>` tag as well.

Let's look at another one: `apple/apple<n><sg>` In this case, the word tested is the lemma itself, apple. apple is a noun (POS) and is singular, hence the `<n><sg>` tags.

## 4.2 - Data Cleaning
[back to Contents](#contents)

The methodology for the data cleaning is as such:
<ol><li>Drop any entries that are incomplete</li>
    <li>Break up the lexeme strings into surface_form, lemma, and modifiers (in a list) for easier analysis later</li>
    </ol>

```python
source_pd_step_1 = source_pd.copy()
source_pd_step_1.dropna(axis=0, inplace=True)
display(source_pd_step_1.describe())

##From the total count below, it seems like no rows were dropped. Awesome.
```

```python
#Defining some functions to apply for us to break up the lexeme strings

def give_surface(lexeme):
    surface, y = lexeme.split("/", 1)
    return surface

def give_lemma(lexeme):
    if ">+" not in lexeme: ## This >+ occurs in German words, where 2 root words are combined to form one surface_form
        surface, y = lexeme.split("/", 1) ## It occurs as surface_form/lemma_1<pos>+lemma_2<pos><tags>
        lemma, y = y.split("<", 1)
        return lemma
    else:
        result = []
        surface, y = lexeme.split("/", 1)
        lemma_1, y = y.split("<", 1)
        result.append(lemma_1)
        while ">+" in y:
            throw, y = y.split(">+", 1)
            new_lemma, y = y.split("<", 1)
            result.append(new_lemma)
        return result
    
def give_modifiers(lexeme):
    result = []
    throw, y = lexeme.split("/", 1)
    throw, y = y.split("<", 1) 
    if ">+" not in y:
        y = y.strip(">")
        tags = y.split("><")
        return tags
    else:
        while ">+" in y:
            new_tag, y = y.split(">+", 1)
            tags = new_tag.split("><")
            result.extend(tags)
            throw, y = y.split("<", 1)
        y = y.strip(">")
        y = y.split("><")
        result.extend(y)
        return result
        
```

```python
source_pd_step_2 = source_pd_step_1.copy()
source_pd_step_2["surface_form"] = source_pd_step_2["lexeme_string"].apply(give_surface)
display(source_pd_step_2.head())
```

```python
source_pd_step_3 = source_pd_step_2.copy()
source_pd_step_3["lemma"] = source_pd_step_3["lexeme_string"].apply(give_lemma)
display(source_pd_step_3.head())
```

```python
source_pd_step_4 = source_pd_step_3.copy()
source_pd_step_4["modifiers"] = source_pd_step_4["lexeme_string"].apply(give_modifiers)
display(source_pd_step_4.head())
```

## 5.2 - Method 2: Difficulty
[back to Contents](#contents)

In this section, we are looking to calculate the difficulty of each language through the following methodology:
<ol><li>Group the dataset by the language being learnt</li>
    <li>Calculate the average difficulty by applying mean on the p_recall scores of all words in that language</li>
    <li>The language with the lowest p_recall score (which is the proportion of answers <b>correct</b>) is thus the hardest language to learn</li></ol>
   <br> 
My hypothesis is that the hardest language to learn is French, based on personal experience. Let's find out.

<a id="learnlang"></a>

### 5.2.1 - Method 2.1: By Learning Language

```python
lang_diff_df = cleaned_df.copy()
lang_diff_df = lang_diff_df.groupby("learning_language")["p_recall"].mean().reset_index(name = "raw_difficulty")
lang_diff_df["new_difficulty"] = lang_diff_df["raw_difficulty"].apply(lambda x: 1-x)
lang_diff_df.sort_values(by = "new_difficulty", ascending = False, inplace = True)
display(lang_diff_df)
```

<div class="alert alert-block alert-success"><h3>Analysis:</h3>
    
From the above, as I randomly hypothesised, we can see that French is the hardest language to learn overall, while Italian is the easiest overall. Meanwhile, on average for Portuguese, Spanish and Italian learners, English has a difficulty score of 0.102123 and ranks third.

### 5.2.2 - Method 2.2: By Language of Learner
[back to Contents](#contents)

The method here slightly differs from the above. To do the calculation this time, we will first be grouping by the language of the learner, then the language being learnt, then repeating the averaging process as above.<br>

## 6.1 - Probe 1: Optimal Frequency
[back to Contents](#contents)

Is there a "best" number of times a word needs to be tested before users started getting 100% consistently? The methodology will be as such:
<ol><li>First calculate the length of the surface_form (or lemma if the surface form is a wildcard e.g. `*sf`) and store it in a new column</li>
    <li>Grab the columns that are useful and sort them according to user_id (first), lexeme_id (second), then timestamp (third) in descending order</li>
    <li>Run through the entries in reverse chronological order and check if the entries fit the following conditions:
        <ul><li>The learner gets 1.0 in the end</li>
            <li>There is a clear turning point where the learner goes from making mistakes to 100% consistently</li>
        </ul>
    </li>
    <li>If the conditions are all met, the information on the history_seen of this particular word for this particular user will be stored in a dictionary, along with the length of the word</li>
    <li>After storing everything in a dictionary, run through the dictionary and for entries that are valid (2 items in list, first item is length, second item is history_seen), process them to calculate a mean value.</li>
    <li>The mean value is the optimal number of times to test a word to fully master it.</li>
    </ol>
    
I have no hypothesis, but I expect the number of times to be around 10?
    

```python
freq_df = cleaned_df.copy()
```

```python
freq_df["length"] = np.where(freq_df["surface_form"]!= "<*sf>", freq_df["surface_form"].apply(len), freq_df["lemma"].apply(len))
freq_df = freq_df[["user_id", "lexeme_id", "timestamp","p_recall","history_seen", "length"]]
```

```python
freq_df.sort_values(by=["user_id", "lexeme_id", "timestamp"], ascending = False, inplace=True)
freq_df.reset_index(drop=True, inplace=True)
display(freq_df.head())
```

```python
massive_dict = {}

for index, row in freq_df.iterrows():
    if row["user_id"] not in massive_dict:
        massive_dict[row["user_id"]] = {}
    if row["lexeme_id"] not in massive_dict[row["user_id"]]:
        if row["p_recall"] != 1.0:
            massive_dict[row["user_id"]][row["lexeme_id"]] = ["no"]
        else:
            massive_dict[row["user_id"]][row["lexeme_id"]] = [row["length"]]
    else:
        if row["p_recall"] == 1.0:
            continue
        else:
            if len(massive_dict[row["user_id"]][row["lexeme_id"]]) == 2:
                continue
            elif massive_dict[row["user_id"]][row["lexeme_id"]][0] =="no":
                continue
            elif len(massive_dict[row["user_id"]][row["lexeme_id"]]) ==1:
                massive_dict[row["user_id"]][row["lexeme_id"]].append(row["history_seen"])
```

<div class="alert alert-block alert-warning"><b>Intermediate Observation</b>
    
This code above takes forever to run, but we finally did it. We now have a dictionary with the following implementation:<br>
<br>

`massive_dict = {
    "user_id_1":{
        "lexeme_id_1": [length, number of times tested before a consistent 100%]
        "lexeme_id_2": [no] #signifies that the learner has yet to get a consistent 100% for this word
        "lexeme_id_3": [length] #occurs when learner has all 100% no "turning point" reflected
        ...
        }
    "user_id_2":{
        ...
        }
        ...
        }`

```python
times = 0
count = 0
for key, values in massive_dict.items():
    for lexeme, record in values.items():
        if len(record) <= 1:
            pass
        else:
            times += record[1]
            count += 1

print (times, count)
print("The optimal number of times before a word is fully learnt is " + str(times/count) + " times.")       
```

<div class= "alert alert-block alert-success"><h3>Analysis:</h3>
<br>
From the above, we can see that the average number of times that users had to encounter a word before consistently getting 100% is <b>17</b> (rounded down).<br>
<br>
However, we must acknowledge the limitations of my method and the dataset. For some words, the data set did not contain all of the users' encounters with the word. The first occurrence of a word for a particular user may already have history_seen of e.g. 8, and from there on it is all 1.0 for p_recall. In those cases, my method ignored them as we cannot be sure that that history_seen is the exact turning point.<br>
<br>
We also excluded data points where even the latest attempt by the user did not get 1.0. As well as words where the user never got 1.0, or got 1.0 for all of his attempts. Those points were excluded as it does now show the turning point, and also due to my method's limitations.<br>
<br>
This may have resulted in my number being higher than what it might actually be. However, I hoped to overcome that limitation through aggregating lots of data, which we managed to achieve, as seen from the 651k data points. As such, I have some confidence in the reliability of my data.

### 6.1.1 Side Probe 1.1: Optimal Frequency wrt word length 
[back to Contents](#contents)

This is an extension of what we did above, out of sheer curiosity. When preparing the implementation for the above problem, the length was also stored in the dictionary so as to enable us in answering this question.

Now that we have the dictionary from above, let us extract out the optimal frequency with respect to the length of the word. Hopefully we can identify a trend and relationship between these 2 factors.

The methodology is:
<ol><li>Run through the items in the dictionary, and for entries that are valid (have 2 items in list, first item being the length and second item being the history_seen), store the appropriate values in another dictionary</li>
    <li>Run through the new dictionary to calculate the average number of times needed for each length of word</li>
    <li>Plot a graph to identify trends</li>
    </ol>
    
My hypothesis is that a longer word will require more testing to master.

```python
length_dict = {}

for key, values in massive_dict.items():
    for lexeme, record in values.items():
        if len(record) <= 1:
            pass
        else:
            if record[0] not in length_dict:
                length_dict[record[0]] = [0,0]
            length_dict[record[0]][0] += record[1]
            length_dict[record[0]][1] += 1
```

```python
for length, lst in length_dict.items():
    length_dict[length].append(lst[0]/lst[1])
```

```python
length_df = pd.DataFrame.from_dict(length_dict, orient = "index",columns = ["sum_times", "num_words", "avg_times"])
length_df.reset_index(inplace=True)
length_df.rename(columns={"index":"length_of_word"}, inplace=True)
length_df.sort_values(by = "length_of_word", inplace=True)
```

```python
fig, ax1 = plt.subplots(figsize=(16,6))
length_df.plot(kind = "bar", x = "length_of_word", y="avg_times", ax = ax1, color = "red")
ax1.set_ylabel("Minimum Number of Times to Test a Word to Achieve Mastery")
ax1.set_xlabel("Length of Word")
ax1.set_title("Minimum Number of Testings before Mastery Against Length of the Word to Master")
plt.xticks(rotation = "horizontal")
plt.show()
```

And we have a very surprising finding! It seems that oddly, a word with longer length actually requires less testing before a learner starts getting it correct consistently, while shorter words, especially words that are 1 character long (common in other languages), actually require more testing! <br>
<br>
This is very likely to be due to the fact that shorter words may actually tend to have flexible and multiple meanings depending on the context that the word is used. Some 1 character words may also be ones that are appended to other words (such as t for t'appelle). In other words, more revision is required to understand all the applications and various contexts.<br>

The good news is the min value of history_seen is 1.0, as seen from the very first describe of the dataset. This means we won't meet any ZeroDivisionErrors!<br>

The methodology is as such:
<ol><li>Calculate the recall score of the history sessions for each record (correct/seen) and store in a new column.</li>

## 7.3 - Finding 3: Hardest Modifier
[back to Contents](#contents)

As an ex-French learner, modifiers were terrifying. Knowing when to use le (masculine) and when to use la (feminine), which tense to use, etc. was simply a nightmare. Thus, I am inspired to find out which modifier(s) is/are the most difficult to learn and understand. An extra finding would be the most common modifiers. <br>

For this implementation, the methodology is as such:
<ol><li>Apply a function on the modifiers to read through them and store their count and p_recall in a dictionary.</li>
    <li>Run through the dictionary and calculate the average p_recall score for each modifier to rank their difficulty</li>
    <li>Identify the hardest modifier</li>
    </ol>

```python
modi_df = cleaned_df.copy()
```

```python
modi_results = {}
def count_modifiers(modi_list, recall):
    for modifier in modi_list:
        if modifier in modi_results:
            modi_results[modifier][0] += recall
            modi_results[modifier][1] += 1
        else:
            modi_results[modifier] = [recall, 1]
```

```python
modi_df[["modifiers", "p_recall"]].apply(lambda x: count_modifiers(*x), axis = 1)
```

```python
## I took  a look through the dictionary, it seems like there are a lot of tags, especially for the French language. 
## Let us first do a ranking of the hardest modifiers, as well as their count. This is because of the possibility of
## outliers as we encountered previously for hardest words.

modi_results_2 = copy.deepcopy(modi_results)

for modifier, lst in modi_results_2.items():
    lst.append(lst[0]/lst[1])
```

```python
modi_df_2 = pd.DataFrame.from_dict(modi_results_2, orient="index", columns = ["total_p_recall", "count", "mean_p_recall"])
modi_df_3 = modi_df_2.copy()
modi_df_3["difficulty_score"] = modi_df_3["mean_p_recall"].apply(lambda x: 1-x)
modi_df_4 = modi_df_3.copy()
modi_df_4.sort_values(by = "difficulty_score", ascending = False, inplace = True)
# display(modi_df_4) ## we can see that there are once again outliers with low count and extremely high difficulty
####################### score as a result.
display(modi_df_4.describe())
```

<div class="alert alert-block alert-warning"><b>Intermediate Observation</b>

Once again, let us fall back on the arbitrary cut off point for the count. This time round, let's use 135 (actually 134.5), the 25 percentile value (same as before).

```python
modi_df_5 = modi_df_4.copy()
modi_df_5 = modi_df_5[modi_df_5["count"]>=135]
display(modi_df_5.head())
```

<div class="alert alert-block alert-warning"><b>Intermediate Observation</b>

One observation at this point is that the modifiers that start with @ seems to be only applicable for French words, which tend to have a lot of tenses like conditional tense etc. Unfortunately, the French modifiers were not mentioned in the GitHub document, hence some decoding on our side will need to be done.<br>
<br>
From just a glance, we can see that 4 out of the 5 most difficult modifiers have @ in them and are related to French. Let us take a closer look.

```python
modi_add = pd.DataFrame({"modi_decoded":["Inanimate", "Conditional Tense", "Compound Past", "Passive", "Past Conditional"]})
modi_df_6 = modi_df_5.head()
modi_df_6.reset_index(inplace=True)
modi_df_6 = modi_df_6.join(modi_add)
display(modi_df_6)

fig, ax1 = plt.subplots(figsize = (16, 6))
modi_df_6.plot(kind = "bar", x = "modi_decoded", y = "difficulty_score", ax= ax1)
ax1.set_ylabel("Difficulty Score")
ax1.set_xlabel("Top 5 Modifiers in Difficulty Score")
ax1.set_title("Difficulty Score of the Top 5 Modifiers")
plt.show()
```

<div class= "alert alert-block alert-success"><h3>Analysis:</h3>

From the above, we can see that the Inanimate modifier is the most difficult one to get right. A simple Google search has revealed that the meaning of Inanimate in linguistics is "belonging to the class of nouns denoting nonliving things". Thus this Inanimate is used on surface_forms that are describing inanimate objects.<br>
<br>
Subsequently, the remaining 4 modifiers have 3 of them being tenses. The tenses are "Conditional Tense", "Compound Past" and "Past Conditional". All these are complicated conjugations in French that I won't be delving too deep into, but just to let everyone know. Lastly, there is the passive modifier, which is rather odd given that normally a passive tone can only be shown through a sentence, but somehow a word can have the same effect. Unfortunately, this is beyond my expertise. <br>
<br>
As such, we can roughly tell that people have difficulty mastering different tenses and understanding words describing inanimate objects.
